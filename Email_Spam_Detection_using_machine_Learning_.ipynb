```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report
```

```python
dataset = pd.read_csv('spam.csv', encoding='latin1')
```

```python
print(dataset)
```

```python
dataset= dataset.where((pd.notnull(dataset)) , '')
```

```python
dataset.loc[dataset['v1'] == 'ham', 'v1',] = 0
dataset.loc[dataset['v1'] == 'spam', 'v1',] = 1
```

```python
X = dataset['v2']

y = dataset['v1']
```

```python
print(X)
```

```python
print(y)
```

```python
 X_train , X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3 , random_state = 42)
```

```python
 print(X.shape)
 print(X_train.shape)
 print(X_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

y_train_features = y_train.astype('int')
y_test_features =  y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
lr = LogisticRegression()
lr.fit(X_train_features , y_train_features)

dtree = DecisionTreeClassifier()
dtree.fit(X_train_features, y_train_features)

svm = SVC(kernel = 'linear')
svm.fit(X_train_features, y_train_features)

nb = MultinomialNB()
nb.fit(X_train_features , y_train_features)

rf = RandomForestClassifier()
rf.fit(X_train_features , y_train_features)

gbr = GradientBoostingClassifier()
gbr.fit(X_train_features, y_train_features)

adb = AdaBoostClassifier()
adb.fit(X_train_features , y_train_features)
```

```python
y_pred1 = lr.predict(X_test_features)
y_pred2 = dtree.predict(X_test_features)
y_pred3 = svm.predict(X_test_features)
y_pred4 = nb.predict(X_test_features)
y_pred5 = rf.predict(X_test_features)
y_pred6 = gbr.predict(X_test_features)
y_pred7 = adb.predict(X_test_features)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_lr = accuracy_score(y_test_features, y_pred1)
report_lr = classification_report(y_test_features, y_pred1)
print("The Accuracy for Logistic Regression is:" , accuracy_lr * 100)
print("The Classification report is\n", report_lr)
```

```python
accuracy_dtree = accuracy_score(y_test_features, y_pred2)
report_dtree = classification_report(y_test_features, y_pred2)
print("The Accuracy for Decision Tree is:" , accuracy_dtree * 100)
print("The Classification report is\n", report_dtree)
```

```python
accuracy_svm = accuracy_score(y_test_features, y_pred3)
report_svm = classification_report(y_test_features, y_pred3)
print("The Accuracy for SUpport vector Machine is:" , accuracy_svm * 100)
print("The Classification report is\n", report_svm)
```

```python
accuracy_nb = accuracy_score(y_test_features, y_pred4)
report_nb = classification_report(y_test_features, y_pred4)
print("The Accuracy for Naive Bayes is:" , accuracy_nb * 100)
print("The Classification report is\n", report_nb)
```

```python
accuracy_rf = accuracy_score(y_test_features, y_pred5)
report_rf = classification_report(y_test_features, y_pred5)
print("The Accuracy for Random Forest is:" , accuracy_rf * 100)
print("The Classification report is\n", report_rf)
```

```python
accuracy_gbr = accuracy_score(y_test_features, y_pred6)
report_gbr = classification_report(y_test_features, y_pred6)
print("The Accuracy for Gradient Boosting Regression is:" , accuracy_gbr * 100)
print("The Classification report is\n", report_gbr)
```

```python
accuracy_adb = accuracy_score(y_test_features, y_pred7)
report_adb = classification_report(y_test_features, y_pred7)
print("The Accuracy for Adaboost is:" , accuracy_adb * 100)
print("The Classification report is\n", report_adb)
```

```python
final_data = pd.DataFrame({'Models' : ['Logistic Regression' ,'Decision Tree','SVM','Naive Bayes','Random Forest','GBR','Adaboost'],
                           'Accuracy' : [accuracy_lr , accuracy_dtree, accuracy_svm, accuracy_nb, accuracy_rf, accuracy_gbr, accuracy_adb]})

print(final_data)
```
