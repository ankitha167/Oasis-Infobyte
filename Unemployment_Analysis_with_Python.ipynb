```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
dataa = pd.read_csv('Unemployment in India.csv')
```

```python
print(dataa)
```

```python
dataa.head(10)
```

```python
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployement Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']
```

```python
print(dataa.isnull())
```

```python
print(dataa.isnull().sum())
```

```python
correlation_matrix = dataa.corr(numeric_only=True)

plt.figure(figsize = (12 , 10))
sns.heatmap(correlation_matrix, annot = True , cmap = 'Pastel1')
plt.title('Correlation Matrix')
plt.show()
```

```python
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployement Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']

plt.figure(figsize=(12,6))
plt.title('Indian Employement')
sns.histplot(x='Estimated Employed', hue='State', data=dataa)
plt.legend(loc='upper left',labels=dataa['State'].unique(), bbox_to_anchor=(1, 1),title='State', ncol=1)
plt.show()
```

```python
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployement Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']


plt.figure(figsize=(12, 6))
plt.title('Indian Employement')
sns.histplot(x='Estimated Unemployement Rate', hue='State', data=dataa)
plt.legend(loc='upper left',labels=dataa['State'].unique(), bbox_to_anchor=(1,1),title='State', ncol=1)
plt.show()
```

```python
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployement Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']

sorted_data = dataa.sort_values(by = 'Estimated Unemployement Rate', ascending=False).head(10)

print(sorted_data)

plt.figure(figsize = (10,6))
sns.barplot(x = 'State' , y = 'Estimated Unemployement Rate' , data = sorted_data)
plt.title('Unemployement rate in order')
plt.xlabel('State')
plt.ylabel('Estimated Unemployement Rate')
plt.show()
```

```python
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployement Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']

sorted_data = dataa.sort_values(by = 'Estimated Employed', ascending=False).head(10)

print(sorted_data)

plt.figure(figsize = (10,6))
sns.barplot(y = 'State' , x = 'Estimated Employed' , data = sorted_data)
plt.title('Employement Rate')
plt.ylabel('State')
plt.xlabel('Estimated Employed')
plt.show()
```

```python
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployement Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']


state_highest_unemployment = dataa.loc[dataa['Estimated Unemployement Rate'].idxmax(), 'State']
print(f"State with the highest unemployment rate: {state_highest_unemployment}")


state_highest_employed = dataa.loc[dataa['Estimated Employed'].idxmax(), 'State']
print(f"State with the highest employed rate: {state_highest_employed}")
```

```python
dataa = pd.read_csv('Unemployment in India.csv')
dataa.columns = ['State', 'Date',
                 'Frequency',
                 'Estimated Unemployment Rate',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate',
                 'Area']
fig = px.bar(dataa, x='State', y='Estimated Unemployment Rate', color='Area',
             title='Unemployment Rate by State', barmode='group')

fig.show()

```

```python
fig = px.pie(dataa, names='State', values='Estimated Unemployment Rate',
             title='Estimated Unemployment Rate Distribution by State')

fig.update_layout(height=700, width=700)

fig.show()
```
