```python

import numpy as np

fuel_type_mapping = {'Petrol': 0, 'Diesel': 1, 'CNG': 2}
dataset['Fuel_Type'] = np.vectorize(fuel_type_mapping.get)(dataset['Fuel_Type'])
print(dataset['Fuel_Type'].unique())
```

```python
dataset['Selling_type'].unique()
```

```python
selling_type_mapping = {'Dealer': 0 , 'Individual' : 1}
dataset['Selling_type'] = np.vectorize(selling_type_mapping.get)(dataset['Selling_type'])
print(dataset['Selling_type'].unique())
```

```python
dataset.head(1)
```

```python
dataset['Transmission'].unique()
```

```python
transmission_mapping = {'Manual':0 , 'Automatic':1}
dataset['Transmission'] = np.vectorize(transmission_mapping.get)(dataset['Transmission'])
print(dataset['Transmission'].unique())
```

```python
dataset.head(5)
```

```python
X = dataset.drop(['Car_Name' , 'Selling_Price'] , axis = 1)
y = dataset['Selling_Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train , y_train)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

gbr = GradientBoostingRegressor()
gbr.fit(X_train , y_train)

xg = XGBRegressor()
xg.fit(X_train, y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = gbr.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming y_test and y_pred1, y_pred2, y_pred3, y_pred4 are your true and predicted values for different models

# R2 Score
score1 = r2_score(y_test, y_pred1)
score2 = r2_score(y_test, y_pred2)
score3 = r2_score(y_test, y_pred3)
score4 = r2_score(y_test, y_pred4)

# Mean Squared Error
mse1 = mean_squared_error(y_test, y_pred1)
mse2 = mean_squared_error(y_test, y_pred2)
mse3 = mean_squared_error(y_test, y_pred3)
mse4 = mean_squared_error(y_test, y_pred4)

# Mean Absolute Error
mae1 = mean_absolute_error(y_test, y_pred1)
mae2 = mean_absolute_error(y_test, y_pred2)
mae3 = mean_absolute_error(y_test, y_pred3)
mae4 = mean_absolute_error(y_test, y_pred4)
```

```python
print(score1, score2, score3, score4)
```

```python
print(mse1, mse2, mse3, mse4)
```

```python
print(mae1, mae2,mae3, mae4)
```

```python
final_data = pd.DataFrame({'Models': ['LR','RF','GBR','XG'] , 'R2_Score' :[score1, score2, score3, score4],
                           'MSE_Score':[mse1,mse2,mse3,mse4],
                           'MAE_Score': [mae1,mae2, mae3, mae4]})
```

```python
print(final_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

Models = final_data['Models']
r2_Score = final_data['R2_Score']

plt.bar(Models, height=r2_Score, color='skyblue')

plt.xlabel('Models')
plt.ylabel('R2 Score')
plt.title('R2 Score Comparison for Different Models')

plt.show()
```

```python
import matplotlib.pyplot as plt

final_data.set_index('Models', inplace=True)


final_data.plot(kind='bar', figsize=(10, 6))
plt.title('Model Comparison')
plt.ylabel('Scores')
plt.xlabel('Models')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))


plt.show()
```
